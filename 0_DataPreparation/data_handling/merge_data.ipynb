```python
import pandas as pd

# Die Dateien importieren (Nur Dateinamen, da sie im gleichen Ordner wie das Notebook liegen)
df_umsatz = pd.read_csv('umsatzdaten_gekuerzt.csv')
df_kiwo = pd.read_csv('kiwo.csv')
df_wetter = pd.read_csv('wetter.csv')

# Sicherstellen, dass alle Datumsspalten als Datumsobjekte erkannt werden
df_umsatz['Datum'] = pd.to_datetime(df_umsatz['Datum'])
df_wetter['Datum'] = pd.to_datetime(df_wetter['Datum'])
df_kiwo['Datum'] = pd.to_datetime(df_kiwo['Datum'])

# Wetterdaten als Grundlage nehmen (alle Datumswerte beibehalten)
df_merged = pd.merge(df_wetter, df_umsatz, on='Datum', how='left')  # Left join für Umsatzdaten

# Jetzt die Kiwo-Daten hinzufügen (auch left join)
df_merged = pd.merge(df_merged, df_kiwo, on='Datum', how='left')

# Die ersten 5 Zeilen anzeigen
print("Die ersten 5 Zeilen der zusammengeführten Daten:")
print(df_merged.head())

# Die Datei im gleichen Ordner speichern (kein vollständiger Pfad nötig)
df_merged.to_csv('merged_data.csv', index=False)

print("Daten wurden erfolgreich gespeichert.")


```

```python
import pandas as pd

# CSV-Dateien einlesen
df_umsatz = pd.read_csv('umsatzdaten_gekuerzt.csv')
df_kiwo = pd.read_csv('kiwo.csv')
df_wetter = pd.read_csv('wetter.csv')

# Datumsspalten umwandeln
df_umsatz['Datum'] = pd.to_datetime(df_umsatz['Datum'])
df_kiwo['Datum'] = pd.to_datetime(df_kiwo['Datum'])
df_wetter['Datum'] = pd.to_datetime(df_wetter['Datum'])

# Daten mergen (wetter als Basis)
df_merged = pd.merge(df_wetter, df_umsatz, on='Datum', how='left')
df_merged = pd.merge(df_merged, df_kiwo, on='Datum', how='left')

# Zahlen formatieren (optional): runde auf 2 Nachkommastellen
df_merged['Temperatur'] = df_merged['Temperatur'].round(2)
df_merged['Umsatz'] = df_merged['Umsatz'].round(2)

# Leere Spalten (nur NaNs) entfernen – falls erwünscht
df_merged = df_merged.dropna(axis=1, how='all')

# Leere Zeilen entfernen (optional, wenn z.B. Datum ohne Daten vorliegt)
df_merged_clean = df_merged.dropna(how='all', subset=df_merged.columns.difference(['Datum']))

# Neue CSV speichern
df_merged_clean.to_csv('merged_data_clean.csv', index=False)

# Vorschau zeigen
print(df_merged_clean.head(10))
```
